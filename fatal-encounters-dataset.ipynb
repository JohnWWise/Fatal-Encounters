```python
%%time
# Libraries
from statsmodels.tsa.arima.model import ARIMA 
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Step 1: Perform a value count by 'Agency or agencies involved' to get the top n agencies
n = 200
top_agencies = form_responses['Agency or agencies involved'].value_counts().nlargest(n).index

# Step 2: Filter the dataframe to include only rows where the agency is in the top n
form_responses_top_n = form_responses[form_responses['Agency or agencies involved'].isin(top_agencies)].copy()

# Step 3: Extract the year from the correct date column
form_responses_top_n['Year'] = pd.to_datetime(
    form_responses_top_n[' Date of injury resulting in death (month/day/year)']
).dt.year

# Step 4: Count the number of deaths for each (Year, Agency) combination
df_aggregated = form_responses_top_n.groupby(['Year', 'Agency or agencies involved']).size().reset_index(name='Deaths')

# Step 5: Create a complete range of years for each agency
all_years = pd.DataFrame({'Year': np.arange(2000, 2022)})

# Step 6: Use a loop to merge this range of years with each agency's data, filling in missing years with zeros
agencies = df_aggregated['Agency or agencies involved'].unique()

filled_dfs = []
for agency in agencies:
    agency_data = df_aggregated[df_aggregated['Agency or agencies involved'] == agency]
    agency_full = all_years.merge(agency_data, on='Year', how='left')
    agency_full['Agency or agencies involved'] = agency
    agency_full['Deaths'] = agency_full['Deaths'].fillna(0)  # Fill missing years with 0 deaths
    filled_dfs.append(agency_full)

# Combine the filled dataframes back into one
df_filled = pd.concat(filled_dfs).reset_index(drop=True)
df = df_filled.copy()

# Convert 'Year' to PeriodIndex
df['Year'] = pd.PeriodIndex(df['Year'], freq='Y')
df.set_index('Year', inplace=True)

# Step 2: Forecasting for each agency and comparing with actual 2021 data
agencies = df['Agency or agencies involved'].unique()
results = []

for agency in agencies:
    agency_data = df[df['Agency or agencies involved'] == agency]
    time_series = agency_data['Deaths']
    
    # Fit an ARIMA model for this agency
    model = ARIMA(time_series, order=(1, 1, 0))
    model_fit = model.fit()
    
    # Forecast for 2021
    forecast = model_fit.forecast(steps=1).iloc[0]
    
    # Retrieve actual 2021 value
    actual = agency_data.loc[agency_data.index == '2021', 'Deaths'].values[0]
    
    # Compare forecast with actual value
    mae = mean_absolute_error([actual], [forecast])
    rmse = np.sqrt(mean_squared_error([actual], [forecast]))
    
    # Calculate sMAPE
    if actual + forecast == 0:
        smape = 0  # Handle cases where both actual and forecast are zero
    else:
        smape = 200 * abs(actual - forecast) / (abs(actual) + abs(forecast))
    
    results.append({
        'Agency': agency,
        'Forecasted 2021': forecast,
        'Actual 2021': actual,
        'MAE': mae,
        'RMSE': rmse,
        'sMAPE': smape
    })

# Convert results to DataFrame for easier viewing
results_df = pd.DataFrame(results)

# Calculate mean metrics
mean_mae = results_df['MAE'].mean()
mean_rmse = results_df['RMSE'].mean()
mean_smape = results_df['sMAPE'].mean()

# Prepare the summary DataFrame
summary_df = pd.DataFrame({
    'Mean MAE': [mean_mae],
    'Mean RMSE': [mean_rmse],
    'Mean sMAPE': [mean_smape]
})

# Display the summary results
print(summary_df)

# Retrain on 2000-2021 data and forecast for 2022, 2023, and 2024
future_predictions = []

for agency in agencies:
    agency_data = df[df['Agency or agencies involved'] == agency]
    time_series = agency_data['Deaths']
    
    # Fit the ARIMA model for this agency on data up to 2021
    model = ARIMA(time_series, order=(1, 1, 0))
    model_fit = model.fit()
    
    # Forecast for 2022, 2023, 2024
    forecast_results = model_fit.get_forecast(steps=3)
    predictions = forecast_results.predicted_mean.round(0).astype(int)
    conf_int = forecast_results.conf_int().round(0).astype(int)
    
    # Set negative confidence intervals to 0
    conf_int[conf_int < 0] = 0
    
    for i, year in enumerate(['2022', '2023', '2024']):
        future_predictions.append({
            'Agency': agency,
            'Year': year,
            'Predicted Deaths': predictions.iloc[i],
            'Lower CI': conf_int.iloc[i, 0],
            'Upper CI': conf_int.iloc[i, 1]
        })

# Convert future_predictions to DataFrame
future_predictions_df = pd.DataFrame(future_predictions)

# Display the future predictions DataFrame
print(future_predictions_df.head())
future_predictions_df.to_csv('./data/out/by_agency_predictions.csv')
```
